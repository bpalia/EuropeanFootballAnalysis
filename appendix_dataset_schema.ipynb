## Data Availability

This appendix part deals with data availability within the **European Soccer Database**. Not all teams have corresponding information from FIFA (11), also some teams have two IDs linked to the same FIFA api ID(3). However, FIFA information is the same for the both IDs as the team is the same.

```python
query = """
        SELECT COUNT (DISTINCT team_api_id) Team_IDs, 
               COUNT (DISTINCT team_fifa_api_id) Team_FIFA_IDs,
               COUNT (DISTINCT team_long_name) Long_Team_Names,
               COUNT (DISTINCT team_short_name) Short_Team_Names
          FROM Team
        """
print("Number of distinct values in Team table:")
display(pd.read_sql(query, conn))

query = """
        SELECT COUNT (DISTINCT team_api_id) Team_IDs, 
               COUNT (DISTINCT team_fifa_api_id) Team_FIFA_IDs,
               COUNT (DISTINCT date) Update_Dates
          FROM Team_Attributes
        """
print("Number of distinct values in Team Attributes table:")
display(pd.read_sql(query, conn))

query = """
        SELECT COUNT (DISTINCT home_team_api_id) Team_IDs
          FROM Match
        """
print("Number of distinct values in Match table:")
display(pd.read_sql(query, conn))

query = """
        SELECT t.*, 
               (SELECT l.name
                  FROM Match m
                 INNER JOIN League l ON m.league_id = l.id
                 WHERE m.home_team_api_id = t.team_api_id
                 LIMIT 1) AS League 
          FROM Team t
         WHERE team_fifa_api_id IS NULL
           OR (SELECT COUNT(*) 
          FROM Team t2
         WHERE t2.team_fifa_api_id = t.team_fifa_api_id) > 1
        """
print("Teams without FIFA information or with multiple names:")
display(pd.read_sql(query, conn))
```

Some players have more than one ID. Also in the Player Attribute table, there is a couple of additional inputs not seen in the Player table.  

```python
query = """
        SELECT COUNT (DISTINCT player_api_id) Player_IDs, 
               COUNT (DISTINCT player_fifa_api_id) Player_FIFA_IDs,
               COUNT (DISTINCT player_name) Player_Names
          FROM Player
        """
print("Number of distinct values in Player table:")
display(pd.read_sql(query, conn))

query = """
        SELECT COUNT (DISTINCT player_api_id) Player_IDs, 
               COUNT (DISTINCT player_fifa_api_id) Player_FIFA_IDs,
               COUNT (DISTINCT date) Update_Dates
          FROM Player_attributes
        """
print("Number of distinct values in Player Attributes table:")
display(pd.read_sql(query, conn))

query = """
        SELECT player_api_id Player_ID, 
               GROUP_CONCAT(DISTINCT player_fifa_api_id) AS Player_FIFA_IDs
          FROM Player_attributes
         GROUP BY player_api_id
        HAVING COUNT(DISTINCT player_fifa_api_id) > 1;
        """
print("Players with multiple FIFA IDs:")
display(pd.read_sql(query, conn))
```

Only *buildUpPlayDribbling* attribute is missing for two thirds of team attribute updates.

Some leagues and seasons have more missing data than others.

```python
query = """
        SELECT  m.match_api_id,
                l.name League, m.season Season, m.stage, m.date,
                m.home_team_api_id, m.away_team_api_id,
                m.home_team_goal, m.away_team_goal,
                m.home_player_1, m.home_player_2, m.home_player_3, 
                m.home_player_4, m.home_player_5, m.home_player_6, 
                m.home_player_7, m.home_player_8, m.home_player_9, 
                m.home_player_10, m.home_player_11, 
                m.away_player_1, m.away_player_2, m.away_player_3, 
                m.away_player_4, m.away_player_5, m.away_player_6, 
                m.away_player_7, m.away_player_8, m.away_player_9, 
                m.away_player_10, m.away_player_11,
                m.home_player_Y1, m.home_player_Y2, m.home_player_Y3, 
                m.home_player_Y4, m.home_player_Y5, m.home_player_Y6, 
                m.home_player_Y7, m.home_player_Y8, m.home_player_Y9, 
                m.home_player_Y10, m.home_player_Y11, 
                m.away_player_Y1, m.away_player_Y2, m.away_player_Y3, 
                m.away_player_Y4, m.away_player_Y5, m.away_player_Y6, 
                m.away_player_Y7, m.away_player_Y8, m.away_player_Y9, 
                m.away_player_Y10, m.away_player_Y11               
          FROM Match m
          LEFT JOIN League l ON m.league_id = l.id                        
        """
matches = pd.read_sql(query, conn)
missing_league = (
    matches.groupby("League").apply(lambda x: x.isna().any(axis=1).sum())
    / matches["League"].value_counts()
)
fig = custom_plots.barplot_counts(
    sr=missing_league,
    percentage=True,
    ylabel="League",
    title="Proportion of Missing Player-Related Data in Different Leagues",
)
plt.show()
missing_league = (
    matches.groupby("Season").apply(lambda x: x.isna().any(axis=1).sum())
    / matches["Season"].value_counts()
)
fig = custom_plots.barplot_counts(
    sr=missing_league,
    percentage=True,
    ylabel="Season",
    title="Proportion of Missing Player-Related Data in Different Seasons",
)
plt.show()
```
